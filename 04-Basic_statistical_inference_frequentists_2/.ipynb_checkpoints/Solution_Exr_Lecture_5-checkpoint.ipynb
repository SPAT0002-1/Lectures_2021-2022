** Exercise: **

What is the maximum likelihood estimator for a sample of data points drawn from a gaussian distribution but each point having a different error $\sigma_i$ ? 


The approach is the same as for homoscedastic errors but replacing $\sigma$ by $\sigma_i$.    
The likelihood is the product of the $p(x \, | \, \mu, \sigma_i)$ associated to each points.   

$$
L = \prod_{i=1}^{N} \frac{1}{\sigma_i \sqrt{2\,\pi}} \, \exp\left[-0.5 \left (\frac{x_i - \mu}{\sigma_i} \right)^2\right] =  \frac{1}{(2\,\pi)^{n/2}} \prod_{i=1}^{N} \frac{1}{\sigma_i} \exp \left (\frac{(x_i - \mu)^2}{2\, \sigma_i^2} \right)
$$


We therefore have :

$$
\ln(L(\mu)) =  -\frac{n}{2}\,\ln(2\pi) - \sum_i \ln(\sigma_i)  - \sum_{i=1}^N \frac{(x_i \,- \, \mu)^2}{2\,\sigma_i^2}, 
$$

So, taking the partial derivative of this likelihood w.r.t. $\mu$, we get:

$$
\hat{\mu} = \frac{\sum_i \, w_i \, x_i}{\sum_i \, w_i}
$$

where $w_i = \sigma^{-2}_i$ . 

Hence the MLE estimator of the mean is simply the weighted arithmetic mean of all measurements. 

````python
import scipy.stats
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
from ipywidgets import interact
````

````python
def plot_dist(x,n, continuous=True, alpha=0.05, distname=''):
    fig, (ax1,ax2) = plt.subplots(1,2,figsize=(10,4.5))
    if continuous:
        density = n.pdf
        title = "Probability Density Function (pdf)"
        plot_symbol = '-'
    else:
        density = n.pmf
        title = "Probability Mass Function"
        plot_symbol = 'bo'
        
    distribution = density(x)
    ax1.plot(x, distribution, plot_symbol, color="blue", label=distname)
    if continuous:
        ta1, ta2 = n.ppf(alpha/2.), n.ppf(1.-alpha/2.)
        xCI = np.linspace(ta1, ta2, 100.)
        ax1.fill_between(xCI, 0, density(xCI), color="blue", alpha=0.6)
    else:  # UNUSED 
        ax1.vlines(x, 0, distribution, colors='b', lw=5, alpha=0.5)
        xCI=x
    ax1.set_title(title)
    
    # Overplot normal distribution
    norm_dist = scipy.stats.norm(0, 1)
    ax1.plot(x, norm_dist.pdf(x), color="red", label='Gauss')
    a1, a2 = norm_dist.ppf(alpha/2.), norm_dist.ppf(1.-alpha/2.)
    xCI_gauss = np.linspace(a1, a2, 100.)
    ax1.vlines(a1, 0, norm_dist.pdf(a1),  color='red')
    ax1.vlines(a2, 0, norm_dist.pdf(a2),  color='red')
    
    cdf_CI = n.cdf(xCI)
    cdf = n.cdf(x)
    ax2.plot(x, cdf, lw=1, color="blue", label=distname)
    ax2.fill_between(xCI, 0, cdf_CI, color="blue", alpha=0.6)
    ax2.set_title("CDF")
    ax2.plot(x, norm_dist.cdf(x), color="red", label='Gauss')
    ax2.hlines(alpha, -3, 3, color='black')
    ax2.hlines(1-alpha, -3, 3, color='black')
    
    return ax1, ax2
````

## III.2.3 Normal distribution with *unknown variance*

**Exercise:**

Let's consider the following sample of 10 points drawn from a normal distribution $N(\mu=20, \sigma=3)$

``` python
mu, sigma = 20, 3.
Npts = 10
norm = scipy.stats.norm(mu, sigma)
x_sample = norm.rvs(Npts)  
```
- Compare the distribution of means when you assume that you know $\sigma$ and when you estimate $\sigma$ directly from the data. (Use directly `scipy.stats.distribution.pdf()`, note that the student's distribution is `scipy.stats.t(dof)`)
- Calculate a 99% confidence interval around the mean (for a stdev estimated from the data) and compare to the results we obtained above. 

````python
# Estimated time (without interact !) 10 minutes
# Comparison of the gaussian and student distrib

def student(k):
    d = scipy.stats.t(k)
    x = np.arange(d.ppf(0.01), d.ppf(0.99), 0.01)
    plot_dist(x, d, continuous=True)
    return

interact(student, k=(1,20,1.) );
````

````python
# Let's try the case of a 1% confidence interval
alpha = 0.01
# Imagine that we have 10 data points drawn from a gaussian of mean 20, and width 3
mu, sigma = 20, 3.
Npts = 10
norm = scipy.stats.norm(mu, sigma)
x_sample = norm.rvs(Npts)   # We generate our fake sample, i.e. draw 10 points from the Gaussian
x_mean = np.mean(x_sample)   # we calculate the mean of our random variable
x_sigma = np.std(x_sample, ddof=1) # Estimated std of the distribution
x_stde = x_sigma/np.sqrt(Npts)   # Estimated Standard error on the mean

# The mean has a distribution N(mu, sigma/sqrt(N))
stde = sigma/np.sqrt(Npts)    # "true" stde
#norm_mean = scipy.stats.norm(x_mean, stde)
#a1, a2 = norm_mean.ppf(alpha/2.), norm_mean.ppf(1-alpha/2.)
norm_mean=scipy.stats.norm(0, 1)
za1, za2 = norm_mean.ppf(alpha/2.), norm_mean.ppf(1-alpha/2.)
a1, a2 = x_mean+za1*stde, x_mean+za2*stde 

print "Stde %.2f and estimator(stde) %.2f" %(stde, x_stde)
t_mean = scipy.stats.t(Npts-1)
t1, t2 = t_mean.ppf(alpha/2.), t_mean.ppf(1-alpha/2.)   # it is centered on a mean 0
at1, at2 = x_mean+t1*x_stde, x_mean+t2*x_stde 

print "z_a/2 = %.2f, t_a/2 = %.2f" %(za1, t1)
print "z_1-a/2 = %.2f, t_1-a/2 = %.2f" %(za2, t2)

print "**************************************"
print "%.2f percents CI around %.2f is [%.2f,%.2f] when sigma is known" %((1-alpha)*100., x_mean, a1, a2)
print "%.2f percents CI around %.2f is [%.2f,%.2f] when sigma is unknown" %((1-alpha)*100., x_mean, at1, at2)
````

** Note that it can happen that CI based on Student's distribution** is *smaller* than the one using the normal one ! The reason is that your estimator of the variance, can sometimes be smaller than the "true" (population) variance. Indeed, the estimator of the variance has its specific distribution (a $chi^2$ distribution).
