### I.5 Gaussian CDF and confidence intervals

- What should be $erf(\infty)$ ? 
- From the CDF definition, it is clear that:

$$
p(a< x < b \,| \,\mu, \sigma) = P(b | \,\mu, \sigma) - P(a | \, \mu, \sigma)
$$

Derive the probability that $p(\mu - M\,\sigma < x < \mu + M \, \sigma)$ for M = 1, 2, 3.


If we write this explicitly, we have:

$$
p(\mu - M\,\sigma< x < \mu - M\,\sigma \,| \,\mu, \sigma) = \frac{1}{2}~\left( 1 + erf \left( \frac{|M|}{\sqrt{2}} \right)  \right) - \frac{1}{2}~\left( 1 - erf \left( \frac{|M|}{\sqrt{2}} \right)  \right) = erf(M/\sqrt{2})
$$

```python
from scipy.special import erf

for M in [1,2,3]:
    print r'p(mu- %.isigma < x < mu+%.isigma) = %.4f ' %(M, M, erf(M/np.sqrt(2)))
```

- What is the interquartile range $q_{75}âˆ’ q_{25}$ for a Gaussian/Normal distribution of standard deviation $\sigma$ ?

For a normal distribution of width $\sigma$ centered on 0, we want to search at which distance $a_1 \, \sigma$ from the mean (0), we have $P(-a_1 \, \sigma \, | \, 0, \sigma) = 0.25$, and $P(a_1 \, \sigma \, | \, 0, \sigma) = 0.75$.   

We have for the first quartile: 

$$
\frac{1}{2}~\left( 1 - erf \left( \frac{a_1 \, \sigma }{\sqrt{2}\sigma} \right)  \right) = 0.25  
$$

$$
0.5 = erf(a_1 \, / \, \sqrt{2}
$$

So $ a_1/\sqrt{2} = erf^{-1}(0.5)$

Now we can calculate the interquartile easily:

$ q_{75} - q_{25} = 2\, a_1 \, \sigma = 2\, \sqrt{2} \, \sigma \, erf^{-1} (0.5) 

We can use python to calculate the value of the inverse of the error function:

```python
from scipy.special import erfinv

'The interquartile range is therefore %.3f  sigma' %(2.*np.sqrt(2)*erfinv(0.5))
```

### I.6.3 Monte-Carlo methods and error estimates:

**Exercise:** 

Calculate and visualise the CDF of the two (sampled) distributions above and estimate from the CDF the median of the distribution. What can you say ?

```python
# create a uniform distribution
uniform_dist = scipy.stats.uniform(0.0001, 1)
x_sample = uniform_dist.rvs(10000)
x = np.linspace(-0.5, 1.5, 10000)
Px = uniform_dist.pdf(x)

# transform the data
y_sample = -2.5*np.log10(x_sample)
y = -2.5*np.log10(x)
Py = 0.4 * np.log(10) * Px * 10**(-0.4*y)

fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(121)
hx = ax.hist(x_sample, 100, histtype='stepfilled', fc='blue', normed=True)
hx_val, hx_bin= hx[0], hx[1]
ax.plot(x, Px, '-k')
ax.set_xlim(-0.2, 1.2)
ax.set_ylim(0, 1.4001)
ax.xaxis.set_major_locator(plt.MaxNLocator(6))
ax.text(0.95, 0.95, r'$p_x(x) = {\rm Uniform}(x)$', va='top', ha='right', transform=ax.transAxes)
ax.set_xlabel('$x$')
ax.set_ylabel('$p_x(x)$')

ax = fig.add_subplot(122)
hy = ax.hist(y_sample, 100, histtype='stepfilled', fc='blue', normed=True)
hy_val, hy_bin = hy[0], hy[1]
ax.plot(y, Py, '-k')
#ax.set_xlim(0.85, 2.9)
ax.xaxis.set_major_locator(plt.MaxNLocator(6))
ax.text(0.95, 0.95, '$y=-2.5\,\log(x)$\n$p_y(y)=p_x(c\,10^{-0.4\,y})$', va='top', ha='right', transform=ax.transAxes)
ax.set_xlabel('$y$')
ax.set_ylabel('$p_y(y)$')

# CDF : Simply sort and rank your sample ! (i.e. 1st element = rank0, 2nd=rank1, 3rd=rank3, ...)
fig, (ax1, ax2) = plt.subplots(figsize=(10, 5), ncols=2)
x_sample.sort()
y_sample.sort()
CDF_Px = np.linspace(0, 1, len(x_sample))
CDF_Py = np.linspace(0, 1, len(y_sample))   # No need to create a second variable ... just for legibility
ax1.plot(x_sample, CDF_Px)
ax2.plot(y_sample, CDF_Py)
ax1.set_ylabel(r'$p(< x)$')
ax2.set_ylabel(r'$p(< y)$')

median_posx = np.searchsorted(CDF_Px, 0.5)
median_posy = np.searchsorted(CDF_Py, 0.5)
print 'Median uniform distribution=', x_sample[median_posx]
print 'Median transformed distribution %.2f vs %.2f' %(y_sample[median_posy], -2.5*np.log10(0.5))

# Alternative more complicated ways to build the cdf 
#dx, dy = np.diff(hx_bin)[0], np.diff(hy_bin)[0]
#CDF_Px = np.cumsum(hx_val * dx)
#CDF_Py = np.cumsum(hy_val * dy)

# Seconf alternative way to create the PDF
#dx, dy = hx_bin[1]-hx_bin[0], hy_bin[1]-hy_bin[0]
#CDF_Px = [sum(hx_val[0:i] * dx) for i in range(len(hx_val))]
#CDF_Py = [sum(hy_val[0:i] * dy) for i in range(len(hy_val))]

#ax1.step(hx_bin[:-1], CDF_Px, where='mid')
#ax2.step(hy_bin[:-1], CDF_Py, where='mid')
#ax1.plot(x, uniform_dist.cdf(x), color='black')
#ax2.set_xlim(0,2)
```

I.6.2 ** Exercise: **
 
Examine the error on the magnitude: $z = -2.5 * log(a) $ using monte-carlo method and compare it to the results one would obtain using the error propogation formula (**THIS is an important exercise**).     
Let's first consider a normalised flux $a = 1$ and a relative error sigma_a = 0.2 ... 

We study $z = -2.5\,\log(a)$

We have from the classical error propagation formula that:

$$
\sigma_z = \left(\frac{\partial z}{\partial a}\right) \sigma_a = \frac{2.5}{\ln(10)} \frac{\sigma_a}{a}  
$$

```python
mu_a, sigma_a = 1., 0.2

# Try to make this interactive ... 

a = scipy.stats.norm(mu_a, sigma_a)

sample_a = a.rvs(10000)
sample_z = -2.5*np.log10(sample_a) 

mu_z, sigma_z = -2.5*np.log10(mu_a), (2.5/np.log(10))*(sigma_a/mu_a)  # error propagation formula
z = scipy.stats.norm(mu_z, sigma_z)

z_hist = plt.hist(sample_z, bins=100, normed=True, color='blue', alpha=0.3, label='MC', range=(-2,2))
x = np.linspace(-5. ,5. , 100)
plt.plot(x, z.pdf(x), color='red', label='Std err')
plt.legend()

sample_mu_z = np.mean(sample_z)
sample_median_z = np.median(sample_z)
sample_sig_z = np.std(sample_z)


scipy.stats.lognorm._fitstart(sigma_z)
lnorm_sig = scipy.stats.lognorm.fit(sample_z )
lnorm = scipy.stats.lognorm(lnorm_sig[0])
plt.plot(x-1., lnorm.pdf(x), color='blue')
plt.vlines(0, 0, 2., lw=2)
plt.xlim(-2, 2)

print "Standard Error: z = %1.3f +- %1.3f " % (mu_z, sigma_z)
print "MC Error: z = %1.3f +- %1.3f vs median = %1.3f" % (sample_mu_z, sample_sig_z, sample_median_z)
print "Fitted lognorm std = %1.3f"% lnorm_sig[0]
```

We see that, when the relative error on the magnitude becomes large, the mean of the magnitude is different from $-2.5 \, \log(\mu)$, the median of the gaussian and lognormal distribution however correspond to each other. We see also that the std deviation poorly characterizes the lognormal width as it is skewed.
